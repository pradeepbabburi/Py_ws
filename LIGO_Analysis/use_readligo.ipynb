```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import readligo as rl
    
# Load the data from a list of files
start = 842656000
stop = 842670000
segList = rl.getsegs(start, stop, 'H1', flag='CBCLOW_CAT2')

# plot a few seconds of each good segment
N = 10000
for(begin, end) in segList:
    # use getstrain() method to load data
    strain, meta, dq = rl.getstrain(begin, end, 'H1')
    # make a plot
    plt.figure()
    ts = meta['dt']
    rel_time = np.arange(0, end-begin, meta['dt'])
    plt.plot(rel_time[0:N], strain[0:N])
    plt.xlabel('Seconds since GPS - ' + str(begin))
plt.show()
```

```python
# Load all LIGO data from a single file
strain, time, chan_dict = rl.loaddata('H-H1_LOSC_4_V1-815411200-4096.hdf5', 'H1')

slice_list = rl.dq_channel_to_seglist(chan_dict['DATA'])
for slice in slice_list:
    time_seg = time[slice]
    strain_seg = strain[slice]
N = 10000
plt.figure()
plt.plot(time_seg[0:N], strain_seg[0:N])
plt.xlabel('Seconds since GPS')
plt.show()
```
